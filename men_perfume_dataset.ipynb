```python
import pandas as pd
```

```python
#Imoprting the csv file 
df= pd.read_csv('ebay_mens_perfume.csv')
df.head()
```

```python
#Renaming the dataset
df= df.rename(columns={'brand':'Perfume_Brand','title':'flavors'})
df.head()
```

```python
#To check whether database contain any NULL values 
df. isnull().any()
```

```python
#To check how many values are Null in Perfume Brand column
df[df['Perfume_Brand'].isna()==1]
```

```python
#To replace null values with Versace
df['Perfume_Brand']=df['Perfume_Brand'].fillna('Versace')
```

```python
#To check how many values are Null in type column
df[df['type'].isna()==1]
```

```python
#To replace null values with Eau de parfum
df['type']=df['type'].fillna('Eau de parfum')
```

```python
#To check how many values are Null in available column
[df['available'].isna()==1]
```

```python
#To drop the available column from dataset
df = df.drop('available', axis=1)
```

```python
#To look into the top 5 rows
df.head()
```

```python
#To check how many values are Null in availableText column
df[df['availableText'].isna()==1]
```

```python
# drop method is used to delete the rows and columns, in this case drop method is used to delect the three rows
df = df.drop([405, 686, 899])
```

```python
#To check how many values are Null in sold column
df[df['sold'].isna()==1]
```

```python
# replace the null values with 0
df['sold']=df['sold'].fillna('0')
```

```python
#To check how many values are Null in lastUpdated column
df[df['lastUpdated'].isna()==1]
```

```python
#Repalcing the null values with a updated time
df['lastUpdated']=df["lastUpdated"].fillna('May 24, 2024 07:56:25 PDT') 
```
